```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import noisereduce as nr

# Load the audio file
rate, data = wavfile.read("aud5.wav")

# Perform noise reduction using noisereduce library
reduced_noise = nr.reduce_noise(y=data, sr=rate)

# Calculate the noise signal
noise_signal = data - reduced_noise

# Perform Fourier Transform on the original signal
fft_data_original = np.fft.fft(data)

# Perform Fourier Transform on the denoised signal
fft_data_denoised = np.fft.fft(reduced_noise)

# Perform Fourier Transform on the noise signal
fft_data_noise = np.fft.fft(noise_signal)

# Plot the original signal in the Fourier domain
plt.figure(figsize=(12, 8))
plt.subplot(3, 2, 1)
plt.plot(np.abs(fft_data_original))
plt.title("Original Signal in Fourier Domain")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")

# Plot the original signal in the time domain
plt.subplot(3, 2, 2)
plt.plot(data)
plt.title("Original Signal in Time Domain")
plt.xlabel("Sample")
plt.ylabel("Amplitude")

# Plot the denoised signal in the Fourier domain
plt.subplot(3, 2, 3)
plt.plot(np.abs(fft_data_denoised))
plt.title("Denoised Signal in Fourier Domain")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")

# Plot the denoised signal in the time domain
plt.subplot(3, 2, 4)
plt.plot(reduced_noise)
plt.title("Denoised Signal in Time Domain")
plt.xlabel("Sample")
plt.ylabel("Amplitude")

# Plot the noise signal in the Fourier domain
plt.subplot(3, 2, 5)
plt.plot(np.abs(fft_data_noise))
plt.title("Noise Signal in Fourier Domain")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")

# Plot the noise signal in the time domain
plt.subplot(3, 2, 6)
plt.plot(noise_signal)
plt.title("Noise Signal in Time Domain")
plt.xlabel("Sample")
plt.ylabel("Amplitude")

# Show the plots
plt.tight_layout()
plt.show()
```
